````python
%load_ext autoreload
%autoreload 2

import glob
import os
from pathlib import Path

from fibsem.structures import FibsemImage
import matplotlib.pyplot as plt
import numpy as np

from pprint import pprint 
from fibsem import calibration
````

````python
# load images
path = "/home/patrick/github/fibsem/scratch/figure"

# show equivalence to default autofocus on standard sample
filenames = sorted(glob.glob(os.path.join(path, "wd_4", "*wd*.tif")))


print(f"{len(filenames)} files found")

wds = []
metrics = []
dog_m = []

fig = plt.figure(figsize=(7.5, 4))

gs = fig.add_gridspec(3,len(filenames)//4)

j = 0
for i, fname in enumerate(filenames):

    img = FibsemImage.load(fname)
    wd = img.metadata.microscope_state.eb_settings.working_distance
    sharpness = calibration._sharpness(img.data)
    dog = calibration._dog(img.data, low=3, high=9)
    
    wds.append(wd)
    metrics.append(sharpness)
    dog_m.append(dog)
    # only add every second image to plot
    if i % 2 == 0:

        if i % 4 == 0:        
            ax1 = fig.add_subplot(gs[0, j])
        else:
            ax1 = fig.add_subplot(gs[1, j])
            j += 1

        ax1.imshow(img.data, cmap="gray")
        ax1.set_title(f"wd: {wd*1e3:.2f} mm")
        ax1.axis("off")


ax2 = fig.add_subplot(gs[2, :])

# standardise metrics
metrics = np.array(metrics) / np.max(metrics)
dog_m = np.array(dog_m) / np.max(dog_m)

wds = np.array(wds) * 1e3
# plot the plot across all the columns of the subplot
ax2.plot(wds, metrics, color="blue", label="Sharpness")
ax2.plot(wds, dog_m, color="green", label="DoG")

# show max sharpness as vertical line
ax2.axvline(wds[np.argmax(metrics)], color="blue", linestyle="dashed", label="Max Sharpness: {:.2f} mm".format(wds[np.argmax(metrics)]))
ax2.axvline(wds[np.argmax(dog_m)], color="green", linestyle="dashed", label="Max DoG: {:.2f} mm".format(wds[np.argmax(dog_m)]))
ax2.axvline(4.0, color="red", linestyle="dashed", label="Default AutoFocus: 4.0mm")
ax2.set_xlabel("Working Distance (mm)")
ax2.set_ylabel("Metric (Normalised))")
ax2.legend(loc="lower left")

# plt.suptitle("Custom AutoFocus Method (Sharpness)")
plt.subplots_adjust(wspace=0.05, hspace=0.05)

# save figure dpi=300
plt.savefig("custom_autofocus.png", dpi=300)

plt.show()

````
