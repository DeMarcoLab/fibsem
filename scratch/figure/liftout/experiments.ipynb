### WAFFLE DETECTION EVALUATION

```python
# select detections from db (waffle method)
SQL_DETECTIONS = """
SELECT e.name, e.method, d.petname, d.stage, d.step, d.feature, d.is_correct, d.dpx_x, d.dpx_y FROM detections d JOIN experiments e ON e.id = d.experiment_id WHERE e.method ='waffle'
"""
df_detections = pd.read_sql(SQL_DETECTIONS, conn)

display(df_detections)

# save to csv
df_detections.to_csv("waffle/waffle_detections.csv")
```

```python
# NB: this is the entire waffle development process, not just the final model

# summary
print(f"{len(df_detections)} detections")
print(f"{len(df_detections['name'].unique())} experiments")
print(f"{len(df_detections['petname'].unique())} lamella")
print(f"{len(df_detections['feature'].unique())} features")
print(f"{len(df_detections['stage'].unique())} stages")
print(f"{len(df_detections['step'].unique())} steps")
print(f"{len(df_detections['method'].unique())} methods")
print("----")


# OVERALL ACCURACY
df_detections_grouped = df_detections.groupby(["method", "is_correct"]).size().reset_index(name='counts')
df_detections_pivot = df_detections.pivot_table(index=['method'], columns=['is_correct'], values='petname', aggfunc='count').reset_index()
df_detections_pivot['accuracy'] = df_detections_pivot['True'] / (df_detections_pivot['True'] + df_detections_pivot['False'])

print("OVERALL ACCURACY")
display(df_detections_pivot)
print("---")

# group by  feature, calculate accuracy
df_detections_grouped = df_detections.groupby(["method", 'feature', "is_correct"]).size().reset_index(name='counts')
df_detections_pivot = df_detections.pivot_table(index=['method', 'feature'], columns=['is_correct'], values='petname', aggfunc='count').reset_index()
df_detections_pivot['accuracy'] = df_detections_pivot['True'] / (df_detections_pivot['True'] + df_detections_pivot['False'])

print("PER FETURE ACCURACY")
display(df_detections_pivot)
print("---")

```
